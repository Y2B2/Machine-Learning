```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx = ny = 4800
```

```python
nxx = np.linspace(-2,1,nx)
nyy = np.linspace(-1.5,1.5,ny)
mx,my=np.meshgrid(nxx,nyy)
c = mx+my*1j
msk = np.ones((nx,ny),dtype=bool)
z=np.zeros((nx,ny),dtype=complex)
for i in range(20):
    np.multiply(z,z,out=z,where=msk)
    np.add(z,c,out=z,where=msk)
    msk=abs(z)<3
```

```python
plt.imshow(1-msk,cmap='grey')
```
